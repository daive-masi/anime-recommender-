```python
import pandas as pd

details = pd.read_csv("../data/raw/details.csv")
stats = pd.read_csv("../data/raw/stats.csv")

# Fusion
anime_df = pd.merge(details, stats, on="mal_id", how="left")

# Nettoyage rapide
anime_df.dropna(subset=["title", "genres", "score"], inplace=True)
anime_df = anime_df[anime_df["score"] > 0]

# Sélection de colonnes pertinentes
cols = ["mal_id", "title", "genres", "themes", "demographics",
        "type", "rating", "score", "members", "year", "popularity", "synopsis"]
anime_df = anime_df[cols]

anime_df.to_csv("../data/processed/anime_dataset.csv", index=False)
print(f"✅ Dataset final : {anime_df.shape[0]} lignes, {anime_df.shape[1]} colonnes")
```
