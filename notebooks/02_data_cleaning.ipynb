```python
import pandas as pd
import ast

df = pd.read_csv("../data/processed/anime_dataset.csv")

# Nettoyer genres, themes, demographics
def clean_list_column(col):
    def parse(x):
        if pd.isna(x):
            return ""
        try:
            items = ast.literal_eval(x)
            if isinstance(items, list):
                return " ".join(items)
        except:
            return str(x)
        return str(x)
    return col.apply(parse)

for c in ["genres", "themes", "demographics"]:
    df[c] = clean_list_column(df[c])

# Nettoyage synopsis (supprimer sources et parenthèses)
df["synopsis"] = df["synopsis"].fillna("").str.replace(r"\(Source:.*?\)", "", regex=True)
df["synopsis"] = df["synopsis"].str.replace("\n", " ").str.strip()

# Supprimer lignes sans score
df = df[df["score"].notna()]

# Sauvegarde du dataset propre
df.to_csv("../data/processed/anime_clean.csv", index=False)
print(f"✅ Nettoyage terminé : {df.shape[0]} lignes")
```
